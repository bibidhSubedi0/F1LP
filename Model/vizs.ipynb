```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("total.csv")
```

```python
race_order = [
    "Bahrain Grand Prix", "Saudi Arabian Grand Prix", "Australian Grand Prix",
    "Japanese Grand Prix", "Chinese Grand Prix", "Miami Grand Prix",
    "Emilia Romagna Grand Prix", "Monaco Grand Prix", "Canadian Grand Prix",
    "Spanish Grand Prix", "Austrian Grand Prix", "British Grand Prix",
    "Hungarian Grand Prix", "Belgian Grand Prix", "Dutch Grand Prix",
    "Italian Grand Prix", "Azerbaijan Grand Prix", "Singapore Grand Prix",
    "United States Grand Prix", "Mexico City Grand Prix", "SÃ£o Paulo Grand Prix",
    "Las Vegas Grand Prix", "Qatar Grand Prix", "Abu Dhabi Grand Prix"
]
```

```python
df["RaceName"] = pd.Categorical(df["RaceName"], categories=race_order, ordered=True)

df["RaceOrderNum"] = df["RaceName"].cat.codes
```

```python
plt.figure(figsize=(14,6))
for driver, sub_df in df.groupby("Driver"):
    sub_df = sub_df.sort_values("RaceName")
    plt.plot(sub_df["RaceName"], sub_df["DriverPointsSoFar"], marker='o', label=driver)

plt.xlabel("Race")
plt.ylabel("Driver Points So Far")
plt.title("Season Points Progression per Driver")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
for driver, sub_df in df.groupby("Driver"):
    sub_df = sub_df.sort_values("RaceName")
    plt.plot(sub_df["RaceName"], sub_df["DriverWinRateThisSeason"], marker='o', label=driver)

plt.xlabel("Race")
plt.ylabel("Win Rate")
plt.title("Driver Win Rate Progression Throughout the Season")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(
    data=df.sort_values("RaceName"),
    x="RaceName",
    y="TeammateComparison",
    hue="Driver",
    marker="o"
)
plt.title("Teammate Comparison Score Over Races")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
for driver, sub_df in df.groupby("Driver"):
    sub_df = sub_df.sort_values("RaceName")
    plt.plot(sub_df["RaceName"], sub_df["AverageQualifyingPosition"], marker='o', label=driver)

plt.xlabel("Race")
plt.ylabel("Average Qualifying Position")
plt.title("Qualifying Position Trend per Driver")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
for driver, sub_df in df.groupby("Driver"):
    sub_df = sub_df.sort_values("RaceName")
    plt.plot(sub_df["RaceName"], sub_df["NextRacePosition"], marker='o', label=driver)

plt.gca().invert_yaxis()  # Lower position = better finish
plt.xlabel("Race")
plt.ylabel("Finishing Position (lower = better)")
plt.title("Finishing Position Across Races")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
